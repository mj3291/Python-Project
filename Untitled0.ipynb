## **Deleting the unwanted columns from the main dataframe**

```python
#deleted the column name 'urban_metric' ,'extra_urban_metric','combined_metric' , 'urban_imperial' , 'extra_urban_imperial',  'combined_imperial'
df_cleaned=df.drop(['tax_band' ,'fuel_cost_12000_miles','fuel_cost_6000_miles','standard_12_months','standard_6_months','first_year_12_months','first_year_6_months','date_of_change','urban_metric' ,'extra_urban_metric','combined_metric' , 'urban_imperial' , 'extra_urban_imperial',  'combined_imperial'],axis=1)  
df_cleaned
```

### **Creating new segments based on the engine cc inside the df_cleaned**

```python
from pandas.core.arrays.interval import NA
#categorizing the cars based on engine cc
engine_capacity_new=df_cleaned['engine_capacity']
my_array_clean = engine_capacity_new[~np.isnan(engine_capacity_new)]
bins=[0,1000,1500,2000,8500]
labels=['microcars','compact cars','mid-size cars','large cars']
df_cleaned['group1']=pd.cut(my_array_clean,bins=bins,labels=labels)
print(df_cleaned['group1'])
```

# ***Cleaning the remaining missing values***

```python
  
import pandas as pd
# Group the dataset by fuel type and calculate the mean emissions for each type
fuel_emissions = new1_df.groupby('fuel_type')['co2','co_emissions', 'nox_emissions'].mean()

# Print the mean emissions for each fuel type
print(fuel_emissions)
fuel_emissions.plot(kind='bar')
plt.xlabel('Fuel Type')
plt.ylabel('Emissions')
plt.title('Mean Emissions by Fuel Type')
plt.show()
```

```python

# Group the data by segment and manufacturer, and calculate the mean CO2 emissions
grouped_data = new1_df.groupby(['group1', 'manufacturer'])['co2'].mean()

# Sort the data by segment and emissions in descending order
sorted_data = grouped_data.reset_index().sort_values(['group1', 'co2'], ascending=[True, False])

# Select the highest emissions for each segment and manufacturer
highest_emissions = sorted_data.groupby('group1').head(1)

# Print the results
print(highest_emissions)
plt.bar(highest_emissions['group1'].astype(str) + '( '+ highest_emissions['manufacturer'].astype(str) + ')' ,highest_emissions['co2'], color='b')
plt.title('Highest CO2 Emissions by Segment and Manufacturer')
plt.xlabel('Segment')
plt.ylabel('CO2 Emissions (g/km)')
plt.xticks(rotation=90)
plt.show()
```

```python
# Group the data by segment and manufacturer, and calculate the mean CO2 emissions
grouped_data_co = new1_df.groupby(['group1', 'manufacturer'])['co_emissions'].mean()

# Sort the data by segment and emissions in descending order
sorted_data_co = grouped_data_co.reset_index().sort_values(['group1','co_emissions'], ascending=[True, False])

# Select the highest emissions for each segment and manufacturer
highest_emissions_co = sorted_data_co.groupby('group1').head(1)

# Print the results
print(highest_emissions_co)
plt.bar(highest_emissions_co['group1'].astype(str) + '( '+ highest_emissions_co['manufacturer'].astype(str) + ')' ,highest_emissions_co['co_emissions'], color='b')
plt.title('Highest CO Emissions by Segment and Manufacturer')
plt.xlabel('Segment')
plt.ylabel('CO Emissions')
plt.xticks(rotation=90)
plt.show()
```
